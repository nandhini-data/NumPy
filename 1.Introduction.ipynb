# <center>NumPy</center>

* **NumPy** is a open source library that is the universal standard for working with numerical data in python.<br>
* It forms the foundation of other libraries like pandas.<br>
* Pandas DataFrame are built on NumPy arrays and can leverage NumPy functions.

The indices,column name and data column are all stored as NumPy arrays.<br>
Pandas add convenient wrappers and functions.
